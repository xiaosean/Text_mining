```python
def search_keyword_text(html_parser, key):
    want_text = re.search(key, html_parser.html())
    want_text_pos = want_text.span()[0]
#     get class name for pyquery search scope
    class_start_idx = html_parser.html()[:want_text_pos].rfind("class=")
    m = re.search('(?<=class=")(.*)"', html_parser.html()[class_start_idx:want_text_pos])
    class_name = "." + m.group()[:-1]
    class_name = class_name.replace(" ", ".")
    return html_parser(class_name).text()
```

```python
def search_keyword_html(html_parser, key):
    want_text = re.search(key, html_parser.html())
    want_text_pos = want_text.span()[0]
#     get class name for pyquery search scope
    class_start_idx = html_parser.html()[:want_text_pos].rfind("class=")
    m = re.search('(?<=class=")(.*)"', html_parser.html()[class_start_idx:want_text_pos])
    class_name = "." + m.group()[:-1]
    class_name = class_name.replace(" ", ".")
    return html_parser(class_name).html()
```
