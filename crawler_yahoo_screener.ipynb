```python
from time import time
import re
from IPython.core.display import display, HTML
from pyquery import PyQuery as pq
import requests
from selenium import webdriver
import pandas as pd
```

```python
crawler_links = [
    "https://news.money-link.com.tw/yahoo/1088_1.html",
    "https://news.money-link.com.tw/yahoo/1088_2.html",
    "https://news.money-link.com.tw/yahoo/1088_3.html"
]
url = crawler_links[1]
```

```python
ip = "http://140.118.127.132:5000"
```

```python
driver = webdriver.PhantomJS(executable_path=r'./phantomjs-2.1.1-macosx/bin/phantomjs')  # PhantomJS
```

```python
driver.get(url)  # 把網址交給瀏覽器 
```

```python
driver.find_element_by_id("addRow").click()
```

```python
driver.page_source
```

```python
web = pq(driver.page_source)
```

```python
csv_columns = []
for td in web(".dataTables_scroll").find(".h3").find("td"):
    col = web(td).find(".dataTables_sizing").text()
    if(col):
        csv_columns.append(col)
#     print(web(td).text())
```

```python
csv_columns
```

```python
df = pd.DataFrame(columns = csv_columns)
df
```

```python
for index, tr in enumerate(web(".dataTables_scrollBody").find("tr")):
    if(index > 0):
#         csv_columns = web(td).text().split(" ")
#         df = pd.DataFrame(columns = csv_columns)
#     csv_columns.append(web(td).text())
#         df.
        row = [web(value).text() for value in web(tr).find("td")]
#         i don't want star value 
        row[-1] = row[-1] [2:]
#         just id 
        row[0] = row[0] [:4]

#         print(web(tr).html())
#         print(len(row))
#         print(len(csv_columns))
#         print(a)
        df = df.append(pd.DataFrame([row], columns = csv_columns))
```

```python
count = 0
for each_row in df.iterrows():
    print(each_row[1]["股名"])
    row = each_row[1]
    data = {'corpid' : row["股名"],
            "strike" : row["成交價 (元)"],
            "gross" : row["營益率"],
            "per" : row["本益比"],
            "yield_rate" : row["殖利率 (%)"],
            "pbr" : row["股價 淨值比"],
            "volatile" : row["波動值"],
            "avg_score" : row["綜合評分"],
            "star" : row["個股星等"]
           }
    r = requests.post(ip + "/put_score", json=data)
```
