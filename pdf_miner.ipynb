```python
from pdfminer.pdfinterp import PDFResourceManager, PDFPageInterpreter
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from pdfminer.pdfpage import PDFPage
from time import time
from io import StringIO
import re
```

# load whole pdf as a text it will return string

```python
def parse_pdf(filename):
    start_time = time()
    rsrcmgr = PDFResourceManager()
    rettxt = StringIO()
    laparams = LAParams()
    # 縦書き文字を横並びで出力する
    laparams.detect_vertical = True
    device = TextConverter(rsrcmgr, rettxt, codec='utf-8', laparams=laparams)
    # 処理するPDFを開く
    fp = open(filename, 'rb')
    interpreter = PDFPageInterpreter(rsrcmgr, device)

    # maxpages：ページ指定（0は全ページ）
    for page in PDFPage.get_pages(fp, pagenos=None, maxpages=0, password=None,caching=True, check_extractable=True):
        interpreter.process_page(page)
    pdf_text = rettxt.getvalue() 
    fp.close()
    device.close()
    rettxt.close()
    print("load %s pdf cost %d time" % (filename, time()-start_time))
    return pdf_text
```

```python
text = parse_pdf("CSR_REPORT.pdf")
```
